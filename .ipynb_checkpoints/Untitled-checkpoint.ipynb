```python
#declaration of beautifulsoup
from bs4 import BeautifulSoup
import requests
import pandas as pd
from urllib.parse import urlparse

directory="india_university"
csvfile="csvfile"

def create_csv_file():
    url = 'http://www.webometrics.info/en/Asia/India'
    
    flag=True
    
    data=[]
    
    page_request=requests.get(url)
    
    search_page=BeautifulSoup(page_request.content,'html.parser')
    
    india_university = pd.DataFrame()#columns=columns)
    
    while flag:
        print("search_page : ", search_page)
        print("\n")
        elements=search_page.find('table',class_='sticky-enabled')
       
        rows = elements.find_all('tr')
        
        for row in rows:
            
            cols = row.find_all('td')
            
            cols = [ele.text.strip() for ele in cols]
            
            data.append([ele for ele in cols if ele])
        #print("ddddddddddddddddddddddddddddddddddddddddddddddd")
        #print(data)
        india_university=india_university.append(data,ignore_index=True)
       
        nextpage1=search_page.find('ul',class_='pager')
        
        if nextpage1: 
            
            next_page = nextpage1.find('li',class_= 'pager-next')
            
            #print("next_page :",next_page)
            #print("\n")
            if next_page:
                
                linkthird=requests.get(url+next_page.a['href'])
                
                search_page = BeautifulSoup(linkthird.content,'html.parser')
                
                #print("link :" ,linkthird)
                #print("\not")
                next_page=False
            else:
                
                break
                
                flag=False
    india_university.to_csv("output.csv", index=False)
    
    #print(india_university)

#this function for executing above functions
if __name__ == '__main__':
    print("process start...........")
    create_csv_file()
    india_universitys=pd.read_csv("output.csv")
    india_universitys.columns = ['ranking', 'world_rank','university','presence_rank','impact_rank','openness_rank','excellence_rank']
    india_universitys.to_csv("output.csv",index=False)
    india_universitys=india_universitys.drop(index=0,axis=0)
    print("successful.......")
```
